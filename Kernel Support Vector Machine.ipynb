**Importing the Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Importing Dataset**


```python
dataset=pd.read_csv('diabetes.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

**Splitting the dataset into Training set and Test set**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

**Matrix of features**

```python
print(x_train)
```

```python
print(x_test)
```

**Dependent variable**

```python
print(y_train)
```

```python
print(y_test)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

**Training the Kernel SVM model on the Training set**

```python
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',random_state=0)
classifier.fit(x_train,y_train)
```

**Predicting a new result**

```python
r=classifier.predict(sc.transform([[1,103,30,38,83,43.3,0.183,33]]))
print(r)
```

**Predicting the test set results**

```python
y_pred=classifier.predict(x_test)
p=np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
print(p)
```

**Visualizing the predicted and actual results of Training set**

```python
plt.scatter(y_train,classifier.predict(x_train),color='red')
plt.plot(y_train,classifier.predict(x_train),color='red')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Predicted vs Actual')
plt.show()
```

**Visualizing the predicted and actual results of Test set**

```python
plt.scatter(y_test,classifier.predict(x_test),color='blue')
plt.plot(y_test,classifier.predict(x_test),color='blue')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Predicted vs Actual')
plt.show()
```

**Making the confusion matrix**

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

**Accuracy**

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

 **F1 Score**

```python
from sklearn.metrics import f1_score
f1=f1_score(y_test,y_pred)
print('F1 Score is',f1)
```
